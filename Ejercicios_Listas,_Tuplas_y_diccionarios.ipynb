# Ejercicios Listas, Tuplas y diccionarios (Core)
Objetivo: Utilizar listas, tuplas y diccionarios para realizar un análisis básico de datos de ventas utilizando Python.

## Carga de Datos
Crea una lista de diccionarios llamada ventas, donde cada diccionario represente una venta.   
Cada venta debe incluir las siguientes claves:  
* «fecha»: una cadena de texto que represente la fecha de la venta (por ejemplo, «2024-01-01»).
* «producto»: una cadena de texto que represente el nombre del producto vendido.
* «cantidad»: un número entero que represente la cantidad de productos vendidos.
* «precio»: un número flotante que represente el precio unitario del producto.

```python
###Se solicito a Copilot un metodo para generar 400 registros de compras de abarrotes ocurridos durante 2025
import random
from datetime import datetime, timedelta

# Lista de productos de abarrotes
productos = [
    "Arroz", "Fideos", "Aceite", "Azúcar", "Sal", "Harina", "Leche",
    "Huevos", "Pan", "Café", "Té", "Galletas", "Atún", "Salsa de tomate",
    "Mantequilla", "Queso", "Yogur", "Cereal", "Avena", "Legumbres",
    "Detergente", "Lavalozas", "Papel higiénico", "Servilletas",
    "Jabón", "Shampoo", "Cepillo de dientes", "Pasta dental"
]

ventas = []

# Fecha inicial del año 2025
fecha_inicio = datetime(2025, 1, 1)

for _ in range(400):
    # Generar una fecha aleatoria dentro de 2025
    dias_random = random.randint(0, 364)
    fecha_venta = fecha_inicio + timedelta(days=dias_random)

    venta = {
        "fecha": fecha_venta.strftime("%Y-%m-%d"),
        "producto": random.choice(productos),
        "cantidad": random.randint(1, 20),
        "precio": round(random.uniform(300.0, 18000.0), 2)
    }

    ventas.append(venta)
```

## Promedio de Precio por Producto:
* Crea un diccionario llamado precios_por_producto donde las claves sean los nombres de los productos y los valores sean tuplas con la suma de los precios de venta de todas las unidades vendidas y la cantidad total de unidades vendidas.  

* Calcula el precio promedio de venta para cada producto utilizando la información de este diccionario.

```python
precios_por_producto = {}
for venta in ventas:
    producto = venta["producto"]
    cantidad = venta["cantidad"]
    precio = venta["precio"]
    if producto not in precios_por_producto:
        precios_por_producto[producto] = (precio * cantidad, cantidad)
    else:
        precios_por_producto[producto] = (precios_por_producto[producto][0] + precio * cantidad, precios_por_producto[producto][1] + cantidad)

#imrpimir valores promedio
for producto, (total_precio, total_cantidad) in precios_por_producto.items():
    precio_promedio = total_precio / total_cantidad
    print(f"Precio Promedio para {producto}: {precio_promedio:.2f}")
```

## Ventas por Día:
Crea un diccionario llamado ingresos_por_dia donde las claves sean las fechas y los valores sean los ingresos totales generados en cada día.  
Utiliza un bucle para calcular los ingresos totales por día y almacenar estos valores en el diccionario.

```python
# Ventas por Día
ingresos_por_dia = {}
for venta in ventas:
    fecha = venta["fecha"]
    precio = venta["precio"]
    cantidad = venta["cantidad"]
    if fecha in ingresos_por_dia:
        ingresos_por_dia[fecha] += precio * cantidad
    else:
        ingresos_por_dia[fecha] = precio * cantidad

#imprimir ingresos por día en orden cronológico
for fecha in sorted(ingresos_por_dia.keys()):
    print(f"{fecha}: {ingresos_por_dia[fecha]:,.2f}")
```

## Representación de Datos:
Crea un diccionario llamado resumen_ventas donde las claves sean los nombres de los productos y los valores sean diccionarios anidados. Cada diccionario anidado debe contener:  
* «cantidad_total»: la cantidad total vendida del producto.
* «ingresos_totales»: los ingresos totales generados por la venta del producto.
* «precio_promedio»: el precio promedio de venta del producto.

```python
resumen_ventas = {}
for venta in ventas:
    producto = venta["producto"]
    precio = venta["precio"]
    cantidad = venta["cantidad"]
    if producto in resumen_ventas:
        resumen_ventas[producto]["cantidad_total"] += cantidad
        resumen_ventas[producto]["ingresos_totales"] += round(precio * cantidad)
        resumen_ventas[producto]["precio_promedio"] = round(resumen_ventas[producto]["ingresos_totales"] / resumen_ventas[producto]["cantidad_total"])
    else:
        resumen_ventas[producto] = {
            "cantidad_total": cantidad,
            "ingresos_totales": round(precio * cantidad),
            "precio_promedio": round(precio)
        }
print("Resumen de Ventas:")
for producto, resumen in resumen_ventas.items():
    print(f"{producto}: Se vendieron {resumen['cantidad_total']} unidades, por un total de {resumen['ingresos_totales']:,} y un precio promedio de {resumen['precio_promedio']:.2f}")
```
